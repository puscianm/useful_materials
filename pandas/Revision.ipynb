## Create pipelines for common transforms

```python
def numeric_to_numbers(input_df, string_to_delete):
    df = input_df.copy()
    for col in df.columns:
        try:
            df[col] = pd.to_numeric(learning_dataset[col].str.replace(string_to_delete,''),errors='raise')
            print("CONVERTED column " + col)
        except:
            #print("NIE ZAMIENIONO kolumny " + col)
            pass
    return df
```

```python
processed_dataset = numeric_to_numbers(learning_dataset, '$')
processed_dataset
```
